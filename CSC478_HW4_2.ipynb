```python
import pandas as pd
import numpy as np
import apriori2 as ap
```

```python
#Part a
```

```python
#playlists
plsts=[]
with open('playlists.txt') as infile:
    for l in infile:
        newl = l.strip().split(' ')
        plsts.append(newl)
final_plst = []
for i in plsts:
    temp = []
    for j in i:
        temp.append(int(j))
    final_plst.append(temp)
```

```python
#song names
song_data = np.genfromtxt('song_names.txt',delimiter='\t',dtype=str,usecols=(1))
```

```python
songs = []
for i in song_data:
    n = i.split('::')
    songs.append(n)
#print(songs)
```

```python
songs = []
for i in song_data:
    temp = []
    n = i.split('::')
    temp.append('  -  '.join(n))
    songs.append(temp)
print(songs)
```

```python
#Part b
```

```python
L, support = ap.apriori(final_plst,.002)
```

```python
print(L)
```

```python
print(support)
```

```python
#Part c
```

```python
rlst = ap.generateRules(L, support, metric='lift', minMetric=20.0)
print(rlst)
```

```python
rlst = ap.generateRules(L, support, metric='confidence', minMetric=0.5)
print(rlst)
```

```python
ap.pntRules(rlst, songs)
```
